```python
#Problem-> Find maximum row or column sum
# Given an NxM 2D array, you need to find out
# which row or column has largest sum (sum of its elements) overall
# amongst all rows and columns.




str1 = input()
str2 = input()

n_m = [int(x) for x in str1.strip().split(" ")]
num_list = [int(x) for x in str2.strip().split(" ")]

num_rows = n_m[0]
num_col = n_m[1]

l1 =[]

k=0

#creating 2-D list
for i in range(num_rows):
    temp_list = []
    for j in range(k,num_col+k):
        
        temp_list.append(num_list[j])
        k = j +1
    l1.append(temp_list)
    
    
#creating list of column sums   
col_sum = []

for i in range(num_col):
    sum1 = 0
    for j in range(num_rows):
        sum1 += l1[j][i]
    
    col_sum.append(sum1)  
    
    
#creating list of row sums

row_sum = []

for i in range(num_rows):
    sum1 = 0
    for j in range(num_col):
#         print(i,j)
        sum1 += l1[i][j]
    row_sum.append(sum1)

    
row_dict = {v: k for v, k in enumerate(row_sum)}
col_dict = {v: k for v, k in enumerate(col_sum)} 


sorted_row_dict = sorted(row_dict.items(), key=lambda x: x[1], reverse = True)
sorted_col_dict = sorted(col_dict.items(), key=lambda x: x[1], reverse = True)

final_dict = {'row': sorted_row_dict[0],'column' : sorted_col_dict[0]}


if final_dict['column'][1] > final_dict['row'][1]:
    print('column', final_dict['column'][0] ,final_dict['column'][1])
    
else:
    print('row', final_dict['row'][0] ,final_dict['row'][1])
```
